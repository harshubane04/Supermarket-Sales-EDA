## Sales Distribution

```python
sns.histplot(df['Sales'],kde=True,bins=30)
plt.title('Total Sales Distribution')
plt.show()
```

### Insight:
- The total sales distribution is right-skewed meaning most transactions are **lower to mid-values**, with fewer high-value purchases. 

## Quantity Distribution

```python
sns.histplot(df['Quantity'],kde=True,bins=10)
plt.title('Quantity Distribution')
plt.show()
```

### Insight:
- Most customers purchase **small quantities** of items. Large quantity purchases occur less frequently.

## Customer Type Distribution

```python
sns.countplot(x='Customer type',data=df,palette='pastel')
plt.title('Customer Type Distribution')
plt.show()
```

### Insight:
- **Member customers are slightly more than Normal customers**. This shows good adoption of the membership program.

## Gender Distribution

```python
sns.countplot(x='Gender',data=df,palette='icefire')
plt.title('Gender Distribution')
plt.show()
```

### Insight:
- **Female customers slightly outnumber Male customers**. Indicates higher shopping frequency among female customers.

## Branch Distribution

```python
sns.countplot(x='Branch',data=df,palette='coolwarm')
plt.title('Branch Distribution')
plt.show()
```

### Insight:
- Transactions are fairly evenly distributed across **Branch Alex, Giza and Cairo.** No branch dominates the dataset. 

## City Distribution

```python
sns.countplot(x='City',data=df,palette='rocket')
plt.title('City Distribution')
plt.show()
```

### Insight:
- Sales are recorded from **Yangon, Naypyitaw and Mandalay.** All cities contribute significantly to the dataset the differences are **not extreme**. 

## Product Line Distribution

```python
sns.countplot(x='Product line',data=df,palette='bright')
plt.xticks(rotation=90)
plt.title('Product Line Distribution')
plt.show()
```

### Insight:
- All six productlines are actively purchased. The distribution is **fairly balanced** across the categories. **Fashion accessories appear slightly more frequent**, while **Health and beauty appears slightly less frequent.**

## Payment Method Distribution

```python
sns.countplot(x='Payment',data=df,palette='Set2')
plt.title('Payment Method Distribution')
plt.show()
```

### Insight:
- Customers use **all three payment methods:** Ewallet, Cash and Credit Card. The distribution is **not heavily skewed** towards a single method. Ewallet and Cash are **slightly more** than Credit card.

## Rating Distribution

```python
sns.histplot(df['Rating'],kde=True,bins=10)
plt.title('Customer Ratings Distribution')
plt.show()
```

### Insight:
- The distribution shows **more customer ratings in the middle to higher range.** Extremely low or extremely high ratings are **less frequent.**

## Correlation Matrix

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='magma')
plt.title('Correlation Matrix')
plt.show()
```

### Insight:
- Strong positive correlations are observed between **Sales, Tax(5%) and Gross income.** Other numerical variables show **weak or moderate correlations.**

## Branch vs Sales

```python
sns.boxplot(x='Branch',y='Sales',data=df,palette='rocket')
plt.title('Total Sales by Branch')
plt.show()
```

### Insight:
- The overall sales distribution across branches appears **comparable,** with no extreme dominance by a single branch.

## Product Line vs Sales

```python
sns.boxplot(x='Sales',y='Product line',data=df,palette='husl')
plt.title('Total Sales by Product Line')
plt.show()
```

### Insight:
- Total sales vary across different product lines. Some product lines show **wider spread and higher variability** in sales value.

## Customer Type vs Sales

```python
sns.boxplot(x='Customer type',y='Sales',data=df,palette='viridis')
plt.title('Total Sales by Customer Type')
plt.show()
```

### Insight:
- Sales distribution for **Member** and **Normal** customers show a noticeable difference. Members show slightly higher sales values, but the difference is **not very large.**

## Gender vs Sales

```python
sns.boxplot(x='Gender',y='Sales',data=df,palette='dark')
plt.title('Total Sales by Gender')
plt.show()
```

### Insight:
- **Female customers have slightly higher sales compared to the male customers.** The sales distribution difference is **not extreme**. This indicates that the female customers contribute more to the sales.

## Quantity vs Sales

```python
sns.scatterplot(x='Quantity',y='Sales',data=df,color='purple')
plt.title('Quantity vs Total Bill')
plt.show()
```

### Insight:
- As the **quantity purchased increases, the total sales amount also increases.** This showa a **clear positive relationship** between quantity and sales. 

## Unit Price vs Quantity

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Quantity',y='Unit price',data=df,color='green')
plt.title('Unit Price Distribution Across Quantity')
plt.show()
```

### Insight:
- The **unit price distributions** are **very similar** across **all quantity levels.** **The median unit price** and **overall price spread** remain **nearly consistent**, indicating that the **quantity purchased** does **not significantly depend** on the **unit price.**

## Outliers Detection --> Sales

```python
sns.boxplot(y='Sales',data=df,color='lavender')
plt.title('Outliers in Total Sales')
plt.show()
```

### Insight:
- A few **high-value transactions** appear as outliers in the sales. These points are valid puchases not data entry errors. They likely represents the **high-priced items.**

## Rating vs Quantity

```python
sns.boxplot(x='Quantity',y='Rating',data=df,palette='colorblind')
plt.title('Rating vs Quantity')
plt.show()
```

### Insight:
- Customer ratings remains **fairly consistent across different quantities purchased.** There is **no clear upward or downward trend** between quantity and rating.

## Rating vs Product Line

```python
sns.boxplot(x='Product line',y='Rating',data=df,palette='tab10')
plt.xticks(rotation=90)
plt.title('Rating by Product Line')
plt.show()
```

### Insight:
- Customer ratings are **distributed similarly across all product lines.** No single product category shows a **consistently higher or lower rating.** 

## Tax vs Sales

```python
sns.scatterplot(x='Tax 5%',y='Sales',data=df,color='green')
plt.title('Tax vs Total Sales')
plt.show()
```

### Insight:
- Tax increases **directly in proportion to sales.** This shows a **clear linear upward pattern** in the plot.
